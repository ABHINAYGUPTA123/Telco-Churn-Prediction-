## Context
"Predict behavior to retain customers. You can analyze all relevant customer data and develop focused customer retention programs." 


### Content

Each row represents a customer, each column contains customer’s attributes described on the column Metadata.

The data set includes information about:

- Customers who left within the last month – the column is called Churn

- Services that each customer has signed up for – phone, multiple lines, internet, online security, online backup, device protection, tech support, and streaming TV and movies

- Customer account information – how long they’ve been a customer, contract, payment method, paperless billing, monthly charges, and total charges

- Demographic info about customers – gender, age range, and if they have partners and dependents

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
# No. of churn(switching to another) and non-churn(remain with same service)
df['Churn'].value_counts()
```

```python
numerical_features = ['tenure','MonthlyCharges']
fig, ax = plt.subplots(1,2 ,figsize=(28,8))
df[df.Churn=='No'][numerical_features].hist(bins=20, color='blue', alpha=0.5, ax=ax)
df[df.Churn=='Yes'][numerical_features].hist(bins=20, color='red', alpha=0.5, ax=ax)

##As per plot we can find that users having monthly charges between 20 and 30 having too much possibilities of staying
##users having Tenure around 0 to 10 haave high possibilities of being churned, users having Tenure around 68 to 72 have high possibilities of being stay
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
df['TotalCharges'].dtypes
sns.displot(df, x='TotalCharges', hue='Churn', element='step')
```

```python
sns.pairplot(df,vars = ['tenure','MonthlyCharges','TotalCharges'], hue="Churn")
```

# Conclusions

# 
1)Churn is significantly higher in Senior Citizens compared to customers who aren't Senior Citizens.

2)Churn is significantly higher in customers that don't have partners compared to those having Partners.

3)Churn is significantly lower in customers having Dependents compared to those not having Dependents.

4)Churn is significantly higher in customers having paperless billing compared to those not having Paperless Billing.

5)Churn is not much different irrespective of customers enjoying a Phone Service.

6)Churn is almost unbiased in terms of gender.

7)Churn is significantly higher in customers paying through electronic checks compared to those having other Payments Methods.

8)Churn is significantly higher in customers having a month-to-month contract and lower in two-years contract.

9)In case of multiline churn difference is very less comparable therefore, it is unbised. 

10)Higher proportion of the customers subscribed Internet Service through Fiber Optic have churned compared to those subscribed the service through DSL. company should focus on DSL service.

11)Churn is significantly higher in customers who haven't subscribed to these serives - Online Security, Online Backup, Device Protection and Tech Support compared to those who subcribed these services.

12)Churn is slightly higher in customers who haven't subscribed streaming TV or movies too. look like unbised.

13)As per plot we can find that users having monthly charges between 20 and 30 having too much possibilities of staying.

14)Customers having Tenure around 0 to 10 have high possibilities of being churned, users having Tenure around 68 to 72 have high possibilities of being stay.

15)The lower the total charges, the higher the churn.
